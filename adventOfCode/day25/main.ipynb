```python
EMPTY_MARKER = '.'
EAST_MARKER = '>'
SOUTH_MARKER = 'v'
```

```python
def get_forward_loc(loc, graph, axis=0):
    forward_loc = np.array(loc)
    forward_loc[axis] += 1
    forward_loc[axis][np.where(forward_loc[axis] == graph.shape[axis])] = 0
    return (forward_loc[0], forward_loc[1])

def step_forward(graph, axis=0):
    if axis==0:
        marker = SOUTH_MARKER
    else:
        marker = EAST_MARKER
    loc = np.where(graph == marker)
    forward_loc = get_forward_loc(loc, graph, axis)
    moveable_idx = np.where(graph[forward_loc] == EMPTY_MARKER)
    
    if len(moveable_idx[0]) == 0:
        return False
        
    # Move all moveable markers forward
    prev_loc = loc[0][moveable_idx], loc[1][moveable_idx]
    next_loc = forward_loc[0][moveable_idx], forward_loc[1][moveable_idx]
    graph[prev_loc] = EMPTY_MARKER
    graph[next_loc] = marker
    return True
      
def cprint(graph):
    out = ""
    for r in graph:
        out += "".join(i for i in r) + "\n"
    print(out)
```
