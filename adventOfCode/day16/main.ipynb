```python
import sys
import numpy as np
import math
import operator
import functools
```

```python
def parsePacket(binary_str):
    version = int(binary_str[0:3], 2)
    type_id = int(binary_str[3:6], 2)
    # print(binary_str)
    # print(f"{version=}, {type_id=}")
    
    # Literal Value
    if type_id == 4:
        value = ""
        rest = binary_str[6:]
        while rest[0] == "1":
            value += rest[1:5]
            rest = rest[5:]
        value += rest[1:5]
        value = int(value, 2)
        # print(f"{value=}")
        return version, rest[5:]
    # operator
    else:
        length_type_id = int(binary_str[6:7], 2)
        if length_type_id == 0:
            # Next 15 bits repr total length
            total_length = int(binary_str[7:7+15], 2)
            # print(f"{total_length=}")
            rest = binary_str[7+15:]
            sub = rest[:total_length]
            sum = version
            while sub.find('1') >= 0:
                v, sub = parsePacket(sub)
                sum += v
            return sum, rest[total_length:]
        else:
            # Next 11 bits repr num of sub-packets
            num_sub_packets = int(binary_str[7:7+11], 2)
            # print(num_sub_packets)
            rest = binary_str[7+11:]
            sum = version
            for i in range(num_sub_packets):
                v, rest = parsePacket(rest)
                sum += v
            return sum, rest
            
def part1(arr):
    for line in arr:
        print(line)
        binary_str = bin(int(line, 16))[2:].zfill(len(line)*4)        
        sum = parsePacket(binary_str)
        # print(f"{sum=}\n")
        
    return
```

```python
def operate(type_id, values):
    if type_id == 0:
        return sum(values)
    elif type_id == 1:
        return functools.reduce(operator.mul, values, 1)
    elif type_id == 2:
        return min(values)
    elif type_id == 3:
        return max(values)
    elif type_id == 5:
        return 1 if values[0] > values[1] else 0
    elif type_id == 6:
        return 1 if values[0] < values[1] else 0
    elif type_id == 7:
        return 1 if values[0] == values[1] else 0
    

def parsePacket2(binary_str):
    version = int(binary_str[0:3], 2)
    type_id = int(binary_str[3:6], 2)
    # print(binary_str)
    # print(f"{version=}, {type_id=}")
    
    # Literal Value
    if type_id == 4:
        value = ""
        rest = binary_str[6:]
        while rest[0] == "1":
            value += rest[1:5]
            rest = rest[5:]
        value += rest[1:5]
        value = int(value, 2)
        # print(f"{value=}")
        return value, rest[5:]
    # operator
    else:
        length_type_id = int(binary_str[6:7], 2)
        if length_type_id == 0:
            # Next 15 bits repr total length
            total_length = int(binary_str[7:7+15], 2)
            # print(f"{total_length=}")
            rest = binary_str[7+15:]
            rest = rest[:total_length]
            
            values = []
            while rest.find('1') >= 0:
                v, rest = parsePacket2(rest)
                values.append(v)
            
            print(values)
            return operate(type_id, values), rest
        else:
            # Next 11 bits repr num of sub-packets
            num_sub_packets = int(binary_str[7:7+11], 2)
            # print(num_sub_packets)
            rest = binary_str[7+11:]
            values = []
            for i in range(num_sub_packets):
                v, rest = parsePacket2(rest)
                values.append(v)
            print(values)
            return operate(type_id, values), rest
        
def part2(arr):
    for line in arr:
        print(line)
        binary_str = bin(int(line, 16))[2:].zfill(len(line)*4)        
        sum = parsePacket2(binary_str)
        print(f"{sum=}\n")
```
